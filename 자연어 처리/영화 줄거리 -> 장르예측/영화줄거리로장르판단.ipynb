```python

from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
import time

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)

# 웹 페이지 이동
driver.get('https://www.kobis.or.kr/kobis/business/mast/mvie/searchMovieList.do#none')

# 조회 버튼 클릭
driver.find_element_by_xpath('//*[@id="searchForm"]/div[1]/div[5]/button[1]').click()

#처리끝날때까지 기다림
time.sleep(3)

dataset = []

for i in range(1, 51):
  print(i, '페이지')
  tbl = driver.find_element_by_css_selector('#content > div.rst_sch > table')
  tbody = tbl.find_element_by_tag_name("tbody")
  rows = tbody.find_elements_by_tag_name("tr")# 테이블에서 줄 추출
  for index, value in enumerate(rows):
    tds=value.find_elements_by_tag_name("td")
    a_tag = tds[0].find_element_by_tag_name("span").find_element_by_tag_name("a")
    title = a_tag.text
    print('title:', title)
    genre = tds[6].find_element_by_tag_name("span").text
    print('genre:', genre)
    a_tag.click() # 제목 클릭해서 상세페이지 로드
    time.sleep(1)
    
    story = driver.find_element_by_css_selector('p.desc_info').text
    print(story)
    data = [title, story, genre]
    dataset.append(data)
    # 상세 페이지 창닫기 버튼 클릭
    driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[1]/a[2]').click()
    time.sleep(1)

  div_paging = driver.find_element_by_css_selector('div.paging')  
  ul = div_paging.find_element_by_tag_name("ul")
  lis = ul.find_elements_by_tag_name("li")
  if i%10==0:
    div_paging.find_elements_by_tag_name("a")[2].click()
  else:
    li = lis[i%10]
    li.find_element_by_tag_name("a").click()
  time.sleep(3)

dataset
```

```python
import pandas as pd

cols=['title', 'story','genre']
df = pd.DataFrame(dataset, columns=cols)
df.to_excel('movie_genre.xls', index=False)
```

```python

from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
import time

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("user-agent=Mozilla/5.0") # 프로그램에서 사이트 접근 막혔을때 사용
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)

# 영화 리뷰 페이지 연결
driver.get('https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=190991&target=')

div = driver.find_element_by_css_selector('div.choice_movie_info')
title = div.find_element_by_tag_name("h5").find_element_by_tag_name("a").text
print(title)
avg_score = driver.find_element_by_css_selector('#old_content > div.choice_movie_box > div.choice_movie_info > table > tbody > tr:nth-child(7) > td > strong').text
print(avg_score)

dataset = []

for i in range(10):
  print(i, '페이지===============')
  table = driver.find_element_by_css_selector('table.list_netizen')
  tbd = table.find_element_by_tag_name("tbody")
  trs = tbd.find_elements_by_tag_name("tr")
  for tr in trs:
    text = tr.find_element_by_css_selector('td.title').text
    arr = text.split('\n')
    print(arr)
    score = arr[2]
    review = arr[3].split('신고')[0]

    print('score:', score)
    print('review:', review)
    writer = tr.find_element_by_css_selector("a.author").text
    print('wrtier:', writer)
    print('==========================')
    data = [review, score, writer]
    dataset.append(data)
  paging = driver.find_element_by_css_selector('div.paging')
  page_num = paging.find_elements_by_tag_name('a')
  # 다음 페이지 이동
  page_num[i+2].click()
  time.sleep(3)

dataset
```

```python
import pandas as pd

sheet1_cols = ['review', 'score', 'writer']
sheet2_cols=['title', 'avg_score']

df1 = pd.DataFrame(dataset, columns=sheet1_cols)
#df1.to_excel('movie_review.xls', index=False, sheet_name='sheet1')
df1
```

```python
movie_info = [[title, avg_score]]
df2 = pd.DataFrame(movie_info, columns=sheet2_cols)
df2
```

```python
with pd.ExcelWriter('movie_review.xls') as writer:  
  df1.to_excel(writer, sheet_name='sheet1', index=False)
  df2.to_excel(writer, sheet_name='sheet2', index=False)
```
